```python
import numpy as np
from sklearn import svm
from sklearn.model_selection import train_test_split
import theano
import os.path
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt
import pandas as pd
import utils
from scipy import stats
from mlxtend.plotting import plot_decision_regions
import matplotlib.pyplot as plt

TRAIN_PATH = 'data/cullpdb+profile_6133_filtered.npy'
TEST_PATH = 'data/cb513+profile_split1.npy.gz'

# TRAIN DATA #
x_in = utils.load_gz(TRAIN_PATH)
x = np.reshape(x_in, (5534, 700, 57))
del x_in
x = x[:, :, :]
labels = x[:, :, 22:30]
amino_acid_residues = np.arange(0, 21)
sequence_profile = np.arange(35, 56)
horizontal_stack = np.hstack((amino_acid_residues, sequence_profile))
x = x[:, :, horizontal_stack]
labels_new = labels.astype('int32')
labels = labels_new.dot(2**np.arange(8)[::-1])
x_new = x.astype('int64')
x = x_new.dot(2**np.arange(42)[::-1])
def nonzero_mode(arr):
    return stats.mode(arr[np.nonzero(arr)]).mode
m = map(nonzero_mode, labels)
labels = np.asarray(m)
labels = labels.flatten()
y_in = utils.load_gz(TEST_PATH)
y = np.reshape(y_in, (514, 700, 57))
del y_in
y = y[:, :, :]
labels_y = y[:, :, 22:30]
y = y[:, :, horizontal_stack]
labels_new_y = labels_y.astype('int32')
labels_y = labels_new_y.dot(2**np.arange(8)[::-1])
y_new = y.astype('int64')
y = y_new.dot(2**np.arange(42)[::-1])
def nonzero_mode(arr):
    return stats.mode(arr[np.nonzero(arr)]).mode
n = map(nonzero_mode, labels_y)
labels_y = np.asarray(n)
labels_y = labels_y.flatten()
X_train, X_test, y_train, y_test =train_test_split(x, labels, test_size = 0.3)
clf = svm.SVC()
clf.fit(X_train, y_train)
accuracy = clf.score(X_test, y_test)
print 'Train accuracy: ',accuracy
prediction = clf.predict(y)
prediction_accuracy = np.sum(((prediction == labels_y).flatten())).astype('float32') /514
print "test data accuracy: ",prediction_accuracy   
```
